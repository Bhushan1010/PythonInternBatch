```python
df = df.sort_values([' height'])
df.head()
```

```python
df.tail()
```

```python
f_height = df[' height'].value_counts()
f_height.plot(kind='bar')
```

```python
f_height
from matplotlib import pyplot as plt
plt.plot(f_height)
plt.hist(f_height)
plt.scatter(f_height[0],f_height[:][1])
plt.show()
```

```python
r = df[' rating'].value_counts()
r.plot(kind='bar')
```

```python
plt.hist(f_height,25)
plt.show()
```
